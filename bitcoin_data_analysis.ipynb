```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plot
import pandas.io.sql as psql
```

```python
#connection params
host='localhost'
user='naziba'
password='postgres'
warehouse_db='data_warehouse'
```

```python
conn = psycopg2.connect(host=host, dbname=warehouse_db, user=user, password=password)
cur=conn.cursor()
```

```python
##Get daily transaction info from the database
select_query="""select date,COUNT(*) as num_transactions,avg(open) as avg_open_price,avg(close) as avg_close_price,max(high) as highest_price,min(low) as lowest_price,sum(volume_btc) as total_volume_btc,sum(volume_currency) as total_volume_currency,avg(weighted_price) as avg_weighted_price from bitstampusd group by date;"""
bitstamp_daily_data=pd.read_sql(select_query, conn)
select_query="""select date,COUNT(*) as num_transactions,avg(open) as avg_open_price,avg(close) as avg_close_price,max(high) as highest_price,min(low) as lowest_price,sum(volume_btc) as total_volume_btc,sum(volume_currency) as total_volume_currency,avg(weighted_price) as avg_weighted_price from coinbaseusd group by date;"""
coinbase_daily_data=pd.read_sql(select_query, conn)
```

```python
bitstamp_daily_data
```

```python
##Get monthly transaction info from the database
select_query="""select date_trunc('month', date) as month, COUNT(*) as num_transactions,avg(open) as avg_open_price,avg(close) as avg_close_price,max(high) as highest_price,min(low) as lowest_price,sum(volume_btc) as total_volume_btc,sum(volume_currency) as total_volume_currency,avg(weighted_price) as avg_weighted_price from bitstampusd group by month;"""
bitstamp_monthly_data=pd.read_sql(select_query, conn)
select_query="""select date_trunc('month', date) as month, COUNT(*) as num_transactions,avg(open) as avg_open_price,avg(close) as avg_close_price,max(high) as highest_price,min(low) as lowest_price,sum(volume_btc) as total_volume_btc,sum(volume_currency) as total_volume_currency,avg(weighted_price) as avg_weighted_price from coinbaseusd group by month;"""
coinbase_monthly_data=pd.read_sql(select_query, conn)
#convert timezone to local to be in sync with postgres 
bitstamp_monthly_data['month']=bitstamp_monthly_data['month'].dt.tz_convert('Asia/Dhaka')
coinbase_monthly_data['month']=coinbase_monthly_data['month'].dt.tz_convert('Asia/Dhaka')
```

```python
bitstamp_monthly_data
```

```python
##Get yearly transaction info from the database
select_query="""select date_trunc('year', date) as year, COUNT(*) as num_transactions,avg(open) as avg_open_price,avg(close) as avg_close_price,max(high) as highest_price,min(low) as lowest_price,sum(volume_btc) as total_volume_btc,sum(volume_currency) as total_volume_currency,avg(weighted_price) as avg_weighted_price from bitstampusd group by year;"""
bitstamp_yearly_data=pd.read_sql(select_query, conn)
select_query="""select date_trunc('year', date) as year, COUNT(*) as num_transactions,avg(open) as avg_open_price,avg(close) as avg_close_price,max(high) as highest_price,min(low) as lowest_price,sum(volume_btc) as total_volume_btc,sum(volume_currency) as total_volume_currency,avg(weighted_price) as avg_weighted_price from coinbaseusd group by year;"""
coinbase_yearly_data=pd.read_sql(select_query, conn)
#convert timezone to local to be in sync with postgres 
bitstamp_yearly_data['year']=bitstamp_yearly_data['year'].dt.tz_convert('Asia/Dhaka')
coinbase_yearly_data['year']=coinbase_yearly_data['year'].dt.tz_convert('Asia/Dhaka')
```

```python
bitstamp_yearly_data
```

Plot number of daily, monthly and yearly transactions

```python
ax = plot.gca()
bitstamp_daily_data.plot(kind='line',x='date',y='num_transactions',color='red',ax=ax,label='bitstamp')
coinbase_daily_data.plot(kind='line',x='date',y='num_transactions',color='blue',ax=ax,label='coinbase')
plot.title('Number of daily transactions')
plot.show()
```

```python
ax = plot.gca()
bitstamp_monthly_data.plot(kind='line',x='month',y='num_transactions',color='red',ax=ax,label='bitstamp')
coinbase_monthly_data.plot(kind='line',x='month',y='num_transactions',color='blue',ax=ax,label='coinbase')
plot.title('Number of monthly transactions')
plot.show()
```

```python
ax = plot.gca()
bitstamp_yearly_data.plot(kind='line',x='year',y='num_transactions',color='red',ax=ax,label='bitstamp')
coinbase_yearly_data.plot(kind='line',x='year',y='num_transactions',color='blue',ax=ax,label='coinbase')
plot.title('Number of yearly transactions')
plot.show()
```

Plot the average price on a daily, monthly and yearly basis

```python
ax = plot.gca()
bitstamp_daily_data.plot(kind='line',x='date',y='avg_weighted_price',color='green',ax=ax,label='bitstamp')
plot.title('Daily Average Price')
coinbase_daily_data.plot(kind='line',x='date',y='avg_weighted_price',color='purple',ax=ax,label='coinbase')
plot.title('Daily Average Price')
plot.show()
```
